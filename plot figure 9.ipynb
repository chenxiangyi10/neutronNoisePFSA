```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w_sizes = [10, 20, 40, 80, 100, 200, 400, 800, 1000]
acc_incore_window_sizes_fs1000 = np.load("acc_incore_window_sizes.npy")
acc_excore_window_sizes_fs1000 = np.load("acc_excore_window_sizes.npy")
acc_incore_window_sizes_fs100 = np.load("acc_incore_window_sizes_fs100.npy")
acc_excore_window_sizes_fs100 = np.load("acc_excore_window_sizes_fs100.npy")
acc_incore_window_sizes_fs10 = np.load("acc_incore_window_sizes_fs10.npy")
acc_excore_window_sizes_fs10 = np.load("acc_excore_window_sizes_fs10.npy")
```

```python
fig, axs = plt.subplots(1, 1, figsize=(5.5, 4))
axs.semilogx(w_sizes, acc_incore_window_sizes_fs1000,  '-o', linestyle="--", c = "black", label="In-core signal, fs = 1000 Hz")
axs.semilogx(w_sizes, acc_excore_window_sizes_fs1000, '-v', c= "black", label="Ex-core signal, fs = 1000 Hz")
axs.semilogx(w_sizes, acc_incore_window_sizes_fs100,  '-x', linestyle="--", c = "red", label="In-core signal, fs = 100 Hz")
axs.semilogx(w_sizes, acc_excore_window_sizes_fs100, '-*', c= "red", label="Ex-core signal, fs = 100 Hz")
axs.semilogx(w_sizes, acc_incore_window_sizes_fs10,  '-8', linestyle="--", c = "blue", label="In-core signal, fs = 10 Hz")
axs.semilogx(w_sizes, acc_excore_window_sizes_fs10, '-p', c= "blue", label="Ex-core signal, fs = 10 Hz")


axs.set_xlabel("Window size [seconds]\n(a)")
axs.set_ylabel("Accuracy")
axs.legend()
plt.ylim([0.3,1.1])
plt.savefig("accs_ds.png", bbox_inches = "tight",
    pad_inches = 0.1, dpi=300)
```
