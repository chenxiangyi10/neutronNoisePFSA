```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w_sizes = [10, 20, 40, 80, 100, 200, 400, 800, 1000]
acc_incore_window_sizes = np.load("acc_incore_window_sizes.npy") # analyzed in PFSA neutron noise analysis1
acc_excore_window_sizes = np.load("acc_excore_window_sizes.npy") # analyzed in PFSA neutron noise analysis2
```

```python
fig, axs = plt.subplots(1, 1, figsize=(3, 3))
axs.semilogx(w_sizes, acc_incore_window_sizes,  '-o', linestyle="--", c = "black", label="In-core signal")
axs.semilogx(w_sizes, acc_excore_window_sizes, '-v', c= "black", label="Ex-core signal")
axs.set_xlabel("Window size [second]")
axs.set_ylabel("Accuracy")
#fig.tight_layout()
axs.legend()
plt.savefig("accs_a.png", bbox_inches = "tight",
    pad_inches = 0, dpi=300)
```
