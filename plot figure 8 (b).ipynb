```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alphabet_sizes = [5, 10, 20, 40, 50, 60, 70, 80, 100, 150, 200, 400, 800] # alphabet sizes
acc_incore_window_sizes = np.load("acc_incore_alphabet_sizes.npy") # analyzed in PFSA neutron noise analysis3
acc_excore_window_sizes = np.load("acc_excore_alphabet_sizes.npy") # analyzed in PFSA neutron noise analysis4
```

```python
fig, axs = plt.subplots(1, 1, figsize=(3, 3))
axs.semilogx(alphabet_sizes, acc_incore_window_sizes,  '-o', linestyle="--", c = "black", label="In-core signal")
axs.semilogx(alphabet_sizes, acc_excore_window_sizes, '-v', c= "black", label="Ex-core signal")
axs.set_xlabel("Alphabet size $|\mathcal{A}|$")
axs.set_ylabel("Accuracy")
#fig.tight_layout()
axs.legend()
plt.savefig("accs_b.png", bbox_inches = "tight",
    pad_inches = 0, dpi=300)
```
